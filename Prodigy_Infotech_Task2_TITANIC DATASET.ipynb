```python
#Creating a function if a person is man,women or child

def man_wom_chi(passenger):
    age=passenger['Age']
    sex=passenger['Sex']
    return 'child' if age < 16 else sex
```

```python
#Creating a new column in the data set
titanic_df['Person']=titanic_df.apply(man_wom_chi,axis=1)
titanic_df.head()
```

```python
# Create a dataset deck_df and we just drop the missing data in Cabin column.
deck_df=titanic_df.dropna(axis=0)
deck_df
```

```python
# Head of the datatset.
deck_df.head()
```

```python
# Create a function to extract level from the Cabin column.
def get_level(passenger):
    cabin=passenger['Cabin']
    return cabin[0]
```

```python
# To avoid settingWithCopyWarning.

pd.options.mode.chained_assignment=None

# Create a new column level in the deck_df column.
deck_df['level']=deck_df.apply(get_level,axis=1)

# Head of the datatset after adding new column level in it.
deck_df.head()
```

```python
# Catplot for finding which deck has most number of passengers.
sns.catplot(x='level',data=deck_df,kind='count',ec='black',lw=2)
plt.title("DECK",fontsize=18)
plt.xlabel("Name of the Deck",fontsize=13)
plt.ylabel("Number of Passengers",fontsize=13)
plt.grid(True)
plt.show()
print('From the above, we can say that most of the passengers belong to deck C on Titanic.')
```

```python
# Where Alone has 0 value assign Yes and where Alone has value greater than 1 assign No.
titanic_df['Alone'].loc[titanic_df['Alone']>0] = 'No'
titanic_df['Alone'].loc[titanic_df['Alone']==0] = 'Yes'
```

```python
# Head of the datatset.
titanic_df.head()
```

```python
# Catplot whether the passenger is travelling with family or without family members.
sns.catplot(data=titanic_df,x='Alone',kind='count',ec='black',lw=2)
plt.title("FAMILY",fontsize=18)
plt.grid(True)
plt.show()
```

```python
# Count-plot about survival of the passengers.
sns.countplot(data = titanic_df,x = 'Survived',ec='black',lw=1,palette='crest')
plt.grid(True)
plt.xlabel("Survived",fontsize=13)
plt.ylabel("Number of Passengers",fontsize=13)
plt.show()
```

```python
# Cat-plot about class of the passengers who survived.
sns.catplot(data = titanic_df,x ='Pclass',y='Survived',kind='point')
plt.grid(True)
plt.ylabel("Survived",fontsize=13)
plt.xlabel("Class of the Passengers",fontsize=13)
plt.show()
```

```python
# Cat-plot about class and gender of the passengers who have less chances of survival.
sns.catplot(data=titanic_df,x='Pclass',y='Survived',kind='point',hue='Person')
plt.grid(True)
plt.ylabel("Survived",fontsize=13)
plt.xlabel("Class of the Passengers",fontsize=13)
plt.show()
print("From the above, we can say that male has less chance of survived than female or child on the Titanic.Among females passengers and child passengers those who are on the class 3 have less chances of survival than that of class 1 and class 2.")
```

```python
# Create lmplot to check the chances of survival of the passengers on the basis of their age.
sns.lmplot(data=titanic_df,x='Age',y='Survived')
plt.ylabel("Survived",fontsize=13)
plt.xlabel("Age of the Passengers",fontsize=13)
plt.grid(True)
plt.show()
```

```python
# Create lmplot to check the chances of survival of the passengers on the basis of their age and class.
sns.lmplot(data=titanic_df,x='Age',y='Survived',hue='Pclass')
plt.ylabel("Survived",fontsize=13)
plt.xlabel("Age of the Passengers",fontsize=13)
plt.grid(True)
plt.show()
print('From the above we can say that, the passengers who have younger in age and of class 1 has greater chances of survival than that of who have older age and of class 3.')
```
